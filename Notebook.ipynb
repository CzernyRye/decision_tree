```python
## Data manipulation
import pandas as pd
from collections import Counter

df = pd.read_csv('./data/hw4-task1-data.tsv', sep='\t',header=None)
```

```python
print 'column idx1:\n'
print Counter(df.loc[:,1].values) , '\n'
print 'column idx6:\n'
print Counter(df.loc[:,6].values) , '\n'
print 'column idx13:\n'
print Counter(df.loc[:,13].values) , '\n'

```

```python
df.loc[:,1].replace('?','Private',inplace = True)
df.loc[:,6].replace('?','Prof-specialty',inplace = True)
df.loc[:,13].replace('?','United-States',inplace = True)
```

```python
print 'column idx1:\n'
print Counter(df.loc[:,1].values) , '\n'
print 'column idx6:\n'
print Counter(df.loc[:,6].values) , '\n'
print 'column idx13:\n'
print Counter(df.loc[:,13].values) , '\n'
```

```python
df.drop(2, axis=1, inplace=True)
```
